Handling Missing Values

```python
data.isna().sum()
```

```python
data['Zipcode']=data['Zipcode'].fillna(data['Zipcode'].mean())
```

```python
data['Latitude']=data['Latitude'].fillna(data['Latitude'].mean())
```

```python
data['Longitude']=data['Longitude'].fillna(data['Longitude'].mean())
```

```python
data['Area of the House from Basement (in Sqft)']=data['Area of the House from Basement (in Sqft)'].fillna(data['Area of the House from Basement (in Sqft)'].median())
```

```python
data['Living Area after Renovation (in Sqft)']=data['Living Area after Renovation (in Sqft)'].fillna(data['Living Area after Renovation (in Sqft)'].median())
```

```python
data['No of Bathrooms']=data['No of Bathrooms'].fillna(data['No of Bathrooms'].median())
```

```python
data['Sale Price']=data['Sale Price'].fillna(data['Sale Price'].mean())
```

```python
data['Flat Area (in Sqft)']=data['Flat Area (in Sqft)'].fillna(data['Flat Area (in Sqft)'].mean())
```

```python
data['Lot Area (in Sqft)']=data['Lot Area (in Sqft)'].fillna(data['Lot Area (in Sqft)'].mean())
```

```python
data['No of Times Visited']=data['No of Times Visited'].fillna(data['No of Times Visited'].mode()[0])
```

Encoding Categorical Variables



```python
data.info()
```

```python
sorted(data['No of Times Visited'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
data= pd.get_dummies(data,columns=['No of Times Visited'],dtype=int,prefix=['notv'])
data
```

```python
sorted(data['Condition of the House'].unique())
```

```python
label_enc=LabelEncoder()
data['Condition of the House']=label_enc.fit_transform(data['Condition of the House'])
data
```

```python
data['Waterfront View']=data["Waterfront View"].str.lower().map({'yes':1,'no':0})
data
```

```python
data.drop('Date House was Sold',axis=1,inplace=True)
data
```

```python
data.info()
```

Scaling Numerical Variables

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
min_max_scaler=MinMaxScaler()
num_cols = ['No of Bedrooms','No of Bathrooms','Flat Area (in Sqft)',
   'Lot Area (in Sqft)','No of Floors','Overall Grade',
   'Area of the House from Basement (in Sqft)','Basement Area (in Sqft)',
   'Age of House (in Years)','Renovated Year','Zipcode','Latitude',
   'Longitude','Living Area after Renovation (in Sqft)',
   'Lot Area after Renovation (in Sqft)']
data[num_cols]=min_max_scaler.fit_transform(data[num_cols])
data
```

```python
data.info()
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split
x=data.drop(columns=['Sale Price'])
y=data['Sale Price']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("shape of X_train:",x_train.shape)
print("shape of X_test:",x_test.shape)
print("shape of y_train:",y_train.shape)
print("shape of y_test:",y_test.shape)
```

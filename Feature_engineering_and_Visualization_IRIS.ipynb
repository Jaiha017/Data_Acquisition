```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,8,9]                                                                                    # plt.xlabel("X-axis")
y = [23,28,21,15,9]                                                                                # plt.ylabel("Y-axis")
                                                                                                   # plt.title("Any suitable title") ls(line style)
plt.plot(x,y,marker = "*" , color = "purple", ls = ":")                                             # axhline means horizontal
plt.ylabel("age")
plt.xlabel("gender")
plt.title("sensus")
plt.plot(y,x)

plt.axhline(y=max(y), color = "green")
plt.axvline(x=min(x), color = "pink")

plt.grid(True)

plt.show()
```

```python
products = ["books", "pen", "pencil", "eraser"]
counts = [200,25,35,15]
plt.bar(products, counts, color = "pink")
plt.show()
```

```python
plt.barh(products, counts, alpha=0.3)
```

```python
plt.pie(counts, labels = products, shadow = True, explode = [1,0,0,0], autopct="%1.1f%%")
```

# Scatter plot

```python
plt.scatter(x,y,color="green", marker="*")
plt.figure(figsize=(4,4))                                                         # can identify outliers
```

```python
plt.scatter(products,counts,color="green")
```

```python
data = np.random.randn(100)
data2 = np.random.uniform(0,5,size=100)
plt.hist(data,bins=20, color ="purple", edgecolor= "pink", alpha = 0.5)
plt.hist(data2,bins = 20, color = "black",edgecolor = "red", alpha = 0.5)
```

```python
fig,ax = plt.subplots(1,3,figsize=(14,10))                                                          # nrows,ncols(rows and columns)
ax[0].set_title("Linechart")
ax[0].plot(x,y)

ax[1].set_title("Barchart")
ax[1].bar(products,counts)

ax[2].set_title("Scatterplot")
ax[2].scatter(x,y)
plt.show()
```

## IRIS dataset

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.shape
```

```python
iris.columns.to_list()
```

```python
sns.scatterplot(data=iris,x="sepal_length",y = "sepal_width", hue = "species")                      # hue for ath category venam
plt.title("sepal length vs sepal width")
```

```python
sns.scatterplot(data=iris,x="petal_length",y = "petal_width", hue = "species")
plt.title("petal length vs petal width")
```

```python
sns.pairplot(data=iris, hue="species")
```

```python
#  to find outliers
sns.boxplot(data=iris,x="species", y="petal_length")
```
